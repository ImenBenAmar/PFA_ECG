# 1.Convertir en niveaux de gris : 

Supprimer la couleur pour travailler plus facilement avec les contrastes.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Charger l'image en niveaux de gris
image_path = "records100_images/00001_lr.png"  # Remplace par ton chemin d'image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

# 2.Appliquer un seuillage :

Transformer l’image en noir et blanc en définissant un seuil pour séparer les signaux du fond.

```python
# Appliquer un seuillage pour extraire les signaux ECG
_, binary_image = cv2.threshold(image, 150, 255, cv2.THRESH_BINARY)
```

# 3.Inverser les couleurs :

Rendre le fond noir et les signaux blancs.

```python
# # Inverser les couleurs (fond noir, signal blanc)
inverted_image = cv2.bitwise_not(binary_image)
```

# 4.Filtrage et amélioration :

Utiliser des techniques pour lisser et améliorer la visibilité des signaux.

# 5.Affichage de l'image résultante:

```python
# Afficher l'image résultante
plt.figure(figsize=(10, 5))
plt.imshow( inverted_image , cmap='gray')
plt.axis('off')
plt.title("ECG avec fond noir et signaux en blanc")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Charger l'image en niveaux de gris
image_path = "ecg pour imen yomna/ali_adeli.jpg"  # Remplace par ton chemin d'image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Appliquer un seuillage pour extraire les signaux ECG
_, binary_image = cv2.threshold(image, 150, 255, cv2.THRESH_BINARY)

# Inverser les couleurs (fond noir, signal blanc)
inverted_image = cv2.bitwise_not(binary_image)


# Afficher l'image résultante
plt.figure(figsize=(10, 5))
plt.imshow(inverted_image, cmap='gray')
plt.axis('off')
plt.title("ECG avec fond noir et signaux en blanc")
plt.show()
```
